## Subsetting and Sorting

```r
# create a data frame for testing
set.seed(13435)
X <- data.frame("var1" = sample(1:5), "var2" = sample(6:10), 
                "var3" = sample(11:15))
X <- X[sample(1:5), ]; X$var2[c(1,3)] = NA
X
```

```r
# subsetting
X[1:2, "var2"]
# dealing with missing value
X[which(X$var2 > 8), ]
# ordering
X[order(X$var1, X$var3), ]
# ordering with plyr
library(plyr)
arrange(X, var1)
arrange(X, desc(var1))
```

```r
# adding rows or columns
X$var4 <- rnorm(5)
X
```

## Summarizing Data

```r
# using another dataset from web
fileUrl <- "https://data.baltimorecity.gov/api/views/k5ry-ef3g/rows.csv?accessType=DOWNLOAD"
download.file(fileUrl, destfile = "./restaurants.csv")
restData <- read.csv("./restaurants.csv")
```

```r
# look at a bit of the data
head(restData, n = 3)
tail(restData, n = 3)
```

```r
# make summary
summary(restData)
str(restData)
```

```r
# quantiles of quantitative variables
quantile(restData$councilDistrict, na.rm = T)
quantile(restData$councilDistrict, probs = c(0.5, 0.75, 0.9))
```

```r
# make table
table(restData$zipCode, useNA = "ifany")
table(restData$councilDistrict, restData$zipCode)
```

```r
# check for missing values
## TRUE=1, while FALSE=0
sum(is.na(restData$councilDistrict))
any(is.na(restData$councilDistrict))
all(restData$zipCode > 0)
```

```r
# row and column sums
colSums(is.na(restData))
all(colSums(is.na(restData) == 0))
```

```r
# values with specific charactersic
## one characteristic
table(restData$zipCode == "21212")
## multiple characteristics
table(restData$zipCode %in% c("21212", "21213"))
restData[restData$zipCode %in% c("21212", "21213"), ]
```

```r
# using another dataset
data(UCBAdmissions)
DF = as.data.frame(UCBAdmissions)
summary(DF)
# making a contingency table (列聯表)
xt <- xtabs(Freq ~ (Gender + Admit), data = DF)
xt
```

```r
# another dataset
# multivariant contingency table
warpbreaks$replicate <- rep(1:9, len = 54)
str(warpbreaks)
## "." means all variables
xt <- xtabs(breaks ~ ., data = warpbreaks)
xt
## flatten all tables
ftable(xt)
```
