gg stands for "Grammar of Graphics"

> “In brief, the grammar tells us that a statistical graphic is a **mapping** from data to **aesthetic** attributes (colour, shape, size) of **geometric** objects (points, lines, bars). The plot may also contain statistical transformations of the data and is drawn on a specific coordinate system” -- ggplot2 book

```r
# we use mpg dataset as an example
library(ggplot2)
str(mpg)
```

# The Basics: qplot()
**Factors** are important for indicatng subsets of the data (if they are to have different properties); they should be labeled

## Scatter plot

```r
qplot(x = displ, y = hwy, data = mpg)
```

```r
# modifying aesthetics
# drv is a factor
qplot(displ, hwy, data = mpg, color = drv)
# or shape=drv
```

```r
# adding a geom (statistic, which is some summary of the data))
## point geom: 圖上那些點
## smooth geom: 那個藍線加上灰色95%信賴區間
qplot(displ, hwy, data = mpg, geom = c("point", "smooth"))
```

## Histograms
1. Specify single variable, and histogram is done.
2. Use the parameter "fill", not "color".

```r
qplot(hwy, data = mpg, fill = drv)
```

## Boxplot
First specify the variable by which you'll split the data, in this case drv, then specify the variable which you want to examine, in this case hwy.

```r
qplot(drv, hwy, data = mpg, geom = "boxplot")
```

```r
qplot(drv, hwy, data = mpg, geom = "boxplot", 
      color = manufacturer)
# Notice that there are still 3 regions of the plot 
# (determined by the factor drv). 
# Each is subdivided into several boxes 
# depicting different manufacturers.
```

## Facets
Subset your data by a factor variable and create separate plots

```r
# ~ 的左邊 indicates the rows of this kind of matrix here
# ~ 的右邊 determines the columns of the panel
qplot(displ, hwy, data = mpg, facets = . ~ drv)
```

```r
qplot(hwy, data = mpg, facets = drv~., bindwidth = 2)
```

# Building plots in layers
Seven basic components of a ggplot2 plot
1. A **data frame**
2. **aesthetic mappings**: how data are mapped to color, size
3. **geoms**: geometric objects like points, lines, shapes.
4. **facets**: for conditional plots.
5. **stats**: sta:s:cal transforma:ons like binning, quantiles, smoothing.
6. **scales**: what scale an aesthe:c map uses (example: male = red, female = blue).
7. **coordinate system**

## The Setup

```r
g <- ggplot(diamonds, aes(x = carat, y = price, color = cut))
```

## The Layer

```r
g + geom_point() + geom_smooth()

# change color from geom_point
g + geom_point(aes(x = carat, y = price), 
               color = "steelblue", size = 4, alpha = 1/2) 
    + geom_smooth()
```

## The Labels

```r
gg <- g + geom_point() + labs(title = "Scatterplot",
                              x = "Carat", y = "Price")
print(gg)
```

## The Theme

```r
# add title and axis text, change legend title
gg1 <- gg + theme(plot.title = element_text(size = 30, face = "bold"),
                 axis.text.x = element_text(size = 15),
                 axis.text.y = element_text(size = 15),
                 axis.title.x = element_text(size = 25),
                 axis.title.y = element_text(size = 25)) + 
    scale_color_discrete(name = "Cut of diamonds")

print(gg1)

gg + theme_bw(base_family = "Times")
```

## The Facets

```r
# columns defined by 'cut'
gg1 + facet_wrap( ~ cut, ncol=3)
```

```r
# row: color, column: cut
gg1 + facet_wrap(color ~ cut)
```

```r
# in a grid
gg1 + facet_grid(color ~ cut)
```
