# Subsetting
### Subsetting Vectors

```r
x <- c("a", "b", "c", "c", "d", "a")
x[2]
x[1:4]
x[x > "a"]
```

### Subsetting Lists

```r
x <- list(foo = 1:4, bar = 0.6)
# [ always returns an object of the same class as the origin;
# can be used to select more than one element (there is one exception)
x[1]
```

```r
# [[ is used to extract elements of a list or a data frame;
# it can only be used to extract a single element and the class
# of the retured object will not necessarily be a list or data frame
x[[1]]
```

```r
# $ is used to extract elements of a list or data frame by name;
# semantics are similar to that of [[
x$bar
```

```r
x[["bar"]]
```

```r
x["bar"]
```

```r
y <- list(foo = 1:4, bar = 0.6, baz = "hello")
y[c(1, 3)]
```

```r
# subsetting nested elements of a list
z <- list(a = list(10, 12, 14), b = c(3.14, 2.81))
z[[c(1, 3)]]
z[[c(2, 1)]]
```

# Dates and Times

```r
# dates are represented by the "Date" class
# dates are stored internally as the number of days since 1970-01-01
x <- as.Date("1970-01-01")
x
class(x)

unclass(x)
```

```r
# times are represented by the "POSIXct" or the "POSIXlt" class
# times are stored internally as the number of seconds since 1970-01-01
# POSIXct is a very large integer
# POSIXlt is a list containing a bunch of useful information
x <- Sys.time()
x

p <- as.POSIXlt(x)
names(unclass(p))
p$sec
```

```r
x <- Sys.time()
x    # already in POSIXct format
unclass(x)
```

```r
# if your dates are written in a different format
datestring <- c("January 10, 2012 10:40", 
                "December 9, 2011 9:10")
x <- strptime(datestring, "%B %d, %Y %H:%M")
x
class(x)
```

```r
# operations
x <- as.POSIXct("2012-10-25 01:00:00")
y <- as.POSIXct("2012-10-25 06:00:00", tz = "GMT")
y-x
```

# Simulation

```r
# generate random Normal variates with a given mean and standard deviation
x <- rnorm(10)
print(x)

y <- rnorm(n = 10, mean = 20, sd = 2)
print(y)

# get an overview
summary(x)
str(x)
```

```r
# set.seed ensures reproduciblity
set.seed(100)
rnorm(5)

rnorm(5)

set.seed(100)
rnorm(5)
```

```r
# generate Poisson data
rpois(n = 10, lambda = 1)

# cumulative distribution
ppois(q = 2, lambda = 2)    # Pr(x <= 2)
ppois(4, 2)                 # Pr(x <= 4)
ppois(6, 2)                 # Pr(x <= 6)
```

```r
set.seed(1)
sample(1:10, 4)
sample(1:10, 4)

sample(letters, 5)

sample(1:10)                  # permutation ()
sample(1:10, replace = TRUE)  # sample w/replacement ()
```
